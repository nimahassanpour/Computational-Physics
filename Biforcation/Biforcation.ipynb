```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def BTM(mu,x):
	if x < 0.5:
		return mu * x
	else:
		return mu - mu * x

mu_min  = 1
mu_max = 2.0

ic = 0.5

mu_sweep = [ ]
x = [ ]


Step = 2000.0
n1 = 1000
n2 = 200


n=(mu_max-mu_min)/Step
for mu in np.arange(mu_min,mu_max,n):
	
	new_ic = ic
	
	for i in range(n1):
		new_ic = BTM(mu,new_ic)
	
	for i in range(n2):
		new_ic = BTM(mu,new_ic)
		mu_sweep.append(mu)
		x.append(new_ic)
```

```python
plt.figure()

plt.title('BTM diagram when intial condition is=%s' % str(ic), fontsize=12)
plt.xlabel(r'$\mu$', fontsize=20)
plt.ylabel(r'$x_n$', fontsize=20)

plt.plot(mu_sweep, x, 'b,')

plt.show()
```
